```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def relu(x):
    return max(0, x)

def leaky_relu(x, negative_slope):
    return max(0, x) + negative_slope * min(0, x)
```

```python
relu_plot = [relu(x) for x in range(-10, 10)]
leaky_relu_plot = [leaky_relu(x, 0.2) for x in range(-10, 10)]
```

```python
sns.lineplot(range(-10, 10), relu_plot)
```

```python
sns.lineplot(range(-10, 10), leaky_relu_plot)
```

```python
len(relu_plot)
len(relu_plot[10:])
len(relu_plot[:10])
```

```python
fig = plt.figure()

plt.plot(range(-10, 1), relu_plot[:11])
plt.plot(range(-10, 1), leaky_relu_plot[:11])
plt.plot(range(0, 10), relu_plot[10:])

plt.legend(['ReLU(x)', 'Leaky ReLU(x)', 'ReLU(x) and Leaky ReLU(x)'])
plt.xlabel("x")
plt.ylabel("Activation Function Output")
plt.title('ReLU and Leaky ReLU Comparison')
```
